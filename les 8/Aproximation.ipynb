```
import numpy as np
import matplotlib.pyplot as plt

# Задаем рандомное распределение точек
xi = np.random.uniform(-8, 8, 200)
yi = 1/(1+np.exp(-np.random.uniform(0.5, 1.5, 200)*xi-np.random.uniform(-1, 1, 200)))
plt.plot(xi,yi, marker="o", ls="")

plt.show()
```

```
from scipy.optimize import fsolve
import math

def equations(p):
    a, b = p
    return ((-2*(yi - 1/(1+np.exp(-a*xi-b)))*xi*(1/(1+np.exp(-a*xi-b)))*(1-1/(1+np.exp(-a*xi-b)))).sum(), (-2*(yi - 1/(1+np.exp(-a*xi-b)))*(1/(1+np.exp(-a*xi-b)))*(1-1/(1+np.exp(-a*xi-b)))).sum())

# Численное решение системы уравнений
a, b =  fsolve(equations, (1, 1))
print (a, b)

x = np.linspace(-8, 8, 200)
yi_new = 1/(1+np.exp(-a*x-b))

plt.plot(xi,yi, marker="o", ls="")
plt.plot(x,yi_new, color='red')


plt.show()
```
