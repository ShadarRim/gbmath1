```
#task 2

from scipy.optimize import fsolve
import math

def equations(p):
    x, y, l = p
    return (4*x+12*y+2*l*x, 64*y + 12*x + 32*l*y, x*x+16*y*y-64)

# Численное решение системы уравнений
for l_i in range(-100,100,1):
  x, y, l =  fsolve(equations, (1, 1, l_i))
  print(l_i, x, y, l)
```

```
#task3,4
def equations(p):
    x, y = p
    return (x*x - y*y + 3*x*y*y*y - 2*x*x*y*y+2*x-3*y-5, 3*y*y*y-2*x*x+2*x*x*x*y-5*x*x*y*y+5)

# Численное решение системы уравнений
for x_i in range(-100,100,1):
  for y_i in range(-100,100,1):
    x, y =  fsolve(equations, (x_i, y_i))
    print(x_i, y_i, x, y)
```

```
# f'(x) = g(x)/2
# g'(x) = 2-2*f(x)
# f(0) = 0
# g(0) = 1
# f(x+dx) = f(x) + f'(x)*dx

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 10000)
f = [0]
g = [1]

d = 0.002
x_root = []
eq = True

for i in range(0, len(x)-1):
  f.append(f[i]+g[i]*(x[i+1]-x[i])/2)
  g.append(g[i]+(2-2*f[i])*(x[i+1]-x[i])/2) 
  if x_root:
    eq = not (x_root[-1]+1)>x[i]
  if ((abs(f[i]-g[i])) < d) and eq:
    x_root.append(x[i])

plt.plot(x,f)
plt.plot(x,g)

print(f'Корни:', *x_root)

plt.show()
```
